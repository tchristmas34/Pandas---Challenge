```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = pd.DataFrame({"Total Players": [len(purchase_data["SN"].unique())]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
number_of_purchases = purchase_data["Price"].count()
total_revenue = purchase_data["Price"].sum()
average_price = total_revenue / number_of_purchases

summary_df = pd.DataFrame({"Number_of_Unique_Items": [unique_items],
                              "Average_Price": [average_price],
                              "Number_of_Purchases": [number_of_purchases],
                              "Total_Revenue": [total_revenue]})

summary_df["Total_Revenue"] = summary_df["Total_Revenue"].map("${:,.2f}".format)
summary_df["Average_Price"] = summary_df["Average_Price"].map("${:.2f}".format)

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
grouping_heroes=purchase_data.groupby('Gender')['SN'].nunique()
total_players = len(purchase_data["SN"].unique())
hero_purchase_data = grouping_heroes.to_frame('Total_Count')
hero_purchase_data['Percentage_of_Players'] = round(hero_purchase_data/total_players*100,2)
hero_purchase_data
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data.groupby('Gender')['Item ID'].count()
purchase_value = purchase_data.groupby('Gender')['Price'].sum()
purchase_analysis_df=purchase_count.to_frame('Purchase_Count')

purchase_analysis_df["Average_Purchase_Price"] = (purchase_value/purchase_count).map("${:.2f}".format)
purchase_analysis_df["Total_Purchase_Value"] = purchase_value.map("${:.2f}".format)
purchase_analysis_df["Average_Total_Purchase_per_Person"] = (purchase_value/grouping_heroes).map("${:.2f}".format)
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9, 14, 19, 24, 29, 34, 39, max(purchase_data['Age'])]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
purchase_data['Age_Bracket'] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
players_count = purchase_data.groupby('Age_Bracket')['SN'].nunique()
age_demos_df = players_count.to_frame('Total_Count').sort_index()
age_demos_df['Percentage_of_Players']=round(age_demos_df['Total_Count']/total_players * 100,2)
age_demos_df
```

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_age_analysis_df = purchase_data.groupby('Age_Bracket')['Purchase ID'].count().to_frame('Purchase_Count').sort_index()

purchase_age_analysis_df['Average_Purchase_Price'] = purchase_data.groupby('Age_Bracket')['Price'].mean().map("${:.2f}".format)

total_purchase_value = purchase_data.groupby('Age_Bracket')['Price'].sum()


purchase_age_analysis_df['Avg_Total_Purchase_per_Person'] = (total_purchase_value /players_count).map("${:.2f}".format)
purchase_age_analysis_df['Total_Purchase_Value'] = total_purchase_value .map("${:.2f}".format)

purchase_age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count=purchase_data.groupby('SN')['Purchase ID'].nunique()
top_spenders_df = purchase_count.to_frame('Purchase_Count')

top_spenders = purchase_data.groupby('SN')['Price'].sum()
top_spenders_df['Average_Purchase_Price'] = (top_spenders/top_spenders_df['Purchase_Count']).map("${:.2f}".format)
top_spenders_df['Total_Purchase_Value'] = top_spenders

value_df = top_spenders_df.sort_values(by='Total_Purchase_Value', ascending=False)
value_df['Total_Purchase_Value'] = value_df['Total_Purchase_Value'].map("${:.2f}".format)
value_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular_items = purchase_data.filter(['Item ID', 'Item Name','Price'], axis=1)
item_amount = most_popular_items.groupby(['Item ID','Item Name'])['Price'].mean()
purchase_count = most_popular_items.groupby('Item ID')['Item Name'].value_counts()
purchase_value = item_amount * purchase_count

most_popular_items_df = purchase_count.to_frame('Purchase_Count')
most_popular_items_df['Item_Price'] = item_amount.map("${:.2f}".format)
most_popular_items_df['Total_Purchase_Value'] = purchase_value
most_popular_items_dfA = most_popular_items_df.sort_values(by="Purchase_Count", ascending=False)
most_popular_items_dfA['Total_Purchase_Value'] = most_popular_items_dfA['Total_Purchase_Value'].map("${:.2f}".format)
most_popular_items_dfA.head()
```
